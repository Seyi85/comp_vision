# Computer Vision Histograms

### Import statements

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Histograms

```python
# Load some images, calculate and plot their color histograms, one color at the time
# try it with different color spaces BGR, HSV, etc
image = r"C:\Users\oolus\OneDrive\Documents\GitHub\comp_vision\03. Histograms, transformations and Bitwise ops\img\day2.jpg"
image1 = cv2.imread(image)
plt.imshow(image1)
```

```python
Hist_b = cv2.calcHist([image1],[0],None,[255],[0,255])
```

```python
plt.plot(Hist_b, color='b')
```

```python
Hist_g = cv2.calcHist([image1],[1],None,[255],[0,255])

plt.plot(Hist_g, color='g')
```

```python
Hist_r = cv2.calcHist([image1],[2],None,[255],[0,255])

plt.plot(Hist_r, color='r')
```

[Hint](# "cv2.calcHist") 


```python
# Now try to plot all 3 channels at the same time, does the histogram give you an idea of what the image looks like?
colours=['b','g','r']
for i,color in enumerate(colours):
    Hist = cv2.calcHist([image1],[i],None,[256],[0,256])
    plt.plot(Hist, color=color)
    plt.xlim(0,256)
plt.show()
```
