## Sharpening

```python
# Create a 3 by 3 sharpening kernel with 
shap_kernel = np.ones((3,3), np.float32)* -1
shap_kernel[1,1] = 9
shap_kernel
```

[Hint](# "All values should sum up to 1 with the highest value in the center of the kernel.") 

```python
# Create a copy of the imagre and apply the kernel to the image unsing cv2.filter2D and display the filtered image
Image = r"C:\Users\oolus\OneDrive\Documents\GitHub\comp_vision\03. Histograms, transformations and Bitwise ops\img\night2.jpg"
Image = cv2.imread(Image)
image = cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)
img= image.copy()
IMG_shape = cv2.filter2D(img, -1,shap_kernel)

plt.imshow(IMG_shape)
plt.show()
```

## Edge detection

```python
# Create a 3 by 3 kernel with numpy to detect edges
edges = np.ones((3,3), np.float32)* -1
edges[1,1] = 8
edges
```

[Hint](# "All values should sum up to 0 with the highest value in the center of the kernel.") 

```python
# Create a copy of the imagre and apply the kernel to the image unsing cv2.filter2D and display the filtered image
img= image.copy()
IMG_edge = cv2.filter2D(img, -1,edges)

plt.imshow(IMG_edge)
plt.show()
```

```python
# Use cv2.Sobel to detect edges in the x orientation
image = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
img= image.copy()
Sobel_x = cv2.Sobel(img,-1,1,0,ksize=5)

plt.imshow(Sobel_x, cmap= 'gray')
plt.show()
```

```python
# Use cv2.Sobel to detect edges in the y orientation
image = cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)
img= image.copy()
Sobel_y = cv2.Sobel(img,-1,0,1,ksize=5)

plt.imshow(Sobel_y, cmap= 'gray')
plt.show()
```

```python
# Use cv2.Sobel and try to detect edges in the x and y orientation
sobel = Sobel_x + Sobel_y
plt.imshow(sobel, cmap= 'gray')
plt.show()
```

```python
# Use cv2.Canny to detect edges
img= image.copy()
edges = cv2.Canny(img,100,200)
plt.imshow(edges, cmap= 'gray')
plt.show()
```
