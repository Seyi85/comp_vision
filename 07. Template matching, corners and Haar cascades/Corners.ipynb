```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image);
```

```python
# We will work with the image above
# Convert it to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
dst = cv2.cornerHarris(image_gray,2,3,0.04)

# Display the image with the detected corners 

img = image.copy()
#dilating the image 
dst = cv2.dilate(dst,None)
#to get optima value
img[dst>0.01*dst.max()]=[255,0,0]


plt.imshow(img);
```

```python
img = cv2.imread('img/chess.jpg')

# Convert img to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)



# Detect the corners using goodFeaturesToTrack
corners = cv2.goodFeaturesToTrack(img_gray, 100,0.01,10)

corners = np.int0(corners)
img_copy = img.copy()

# Draw the corners on top of img and show it on screen (make sure the corners are visible!)

for corner in corners:
    x,y = corner[0]
    cv2.circle(img_copy, (x,y), 10, (255,0,0), -1)
    
plt.imshow(img_copy);
```
