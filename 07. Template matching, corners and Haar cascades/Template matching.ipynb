```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Template Matching

```python
# Load img/WandoScene.jpg  as the main image

# Load img/waldo.jpg as the template

# Corvert the images to grayscale

# Find the template in the image

# Draw a rectagle around the the detected match
```

```python
# Load img/WandoScene.jpg  as the main image

wando = cv2.imread('img/WandoScene.jpg')

plt.imshow(wando)
```

```python
wando_RGB = cv2.cvtColor(wando, cv2.COLOR_BGR2RGB)

plt.imshow(wando_RGB)
```

```python
# Load img/waldo.jpg as the template
template = cv2.imread('img/waldo.jpg')
temp_RGB = cv2.cvtColor(template, cv2.COLOR_BGR2RGB)
plt.imshow(temp_RGB)
```

```python
# Corvert the images to grayscale

wando_gray = cv2.cvtColor(wando_RGB, cv2.COLOR_BGR2GRAY)
template_gray = cv2.cvtColor(temp_RGB, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(wando_gray, cmap='gray');
```

```python
plt.imshow(template_gray, cmap='gray');
```

```python
# Draw a rectagle around the the detected match
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
wando_copy = wando.copy()
for m in methods:
    method = eval(m)
    
    wando_gray_copy = wando_gray.copy()
    res = cv2.matchTemplate(wando_gray_copy, template_gray, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    h,w,c = template.shape
    bottom_right = (top_left[0]+50, top_left[1]+50)  
    cv2.rectangle(wando_copy, top_left, bottom_right, (0,0,255), 10)
    
    plt.figure(figsize=(15,15))
    plt.subplot(121)
    plt.imshow(res, cmap='gray')
    plt.title('HEATMAP OF TEMPLATE MATCHING')
    
    #plt.figure(figsize=(10,8))
    plt.subplot(122)
    plt.imshow(wando_copy[:,:,::-1])
    plt.title('DETECTION OF TEMPLATE')
    plt.suptitle(m)
    
    plt.show()
    print()
    print()
```
